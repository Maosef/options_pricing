```python
##STEP 1##
import requests
import numpy
from bs4 import BeautifulSoup
import pandas as pd

##First 2 lines get the data for us
url = "https://www.spaceweatherlive.com/en/solar-activity/top-50-solar-flares"
r = requests.get( url )
soup = BeautifulSoup( r.content, 'html.parser')
html_data = soup.prettify()
#helps us find the table
html_table = soup.find("table")
#print(html_data)
#print(html_table)
#Puts the table into a dataframe
df=pd.read_html( "https://www.spaceweatherlive.com/en/solar-activity/top-50-solar-flares")[0]
#Changes the column names
df.columns=['rank', 'x_classification', 'date', 'region', 'start_time', 'maximum_time', 'end_time', 'movie']
```

```python
df
```

```python
##STEP 2##

#We fix missing values, and remove a column
import datetime as dt
df=df.replace('-','NaN')
df=df.drop('movie',axis=1)

#We iterate through the rows and combine the date and time into one timestampe
for row, col in df.iterrows():
    df.loc[:,'start_datetime'] = (pd.to_datetime(df['date'] + ' ' + df['start_time']))
    df.loc[:,'max_datetime'] = (pd.to_datetime(df['date'] + ' ' + df['maximum_time']))
    df.loc[:,'end_datetime'] = (pd.to_datetime(df['date'] + ' ' + df['end_time']))

#We drop our old columns
df = df.drop('start_time', axis=1)
df = df.drop('maximum_time', axis=1)
df = df.drop('end_time', axis=1)
```

```python
##STEP 3##

#The first 2 lines get the data for us
url2='http://cdaw.gsfc.nasa.gov/CME_list/radio/waves_type2.html'
r2 = requests.get(url2)
soup2 = BeautifulSoup(r2.content,'html.parser')
#The find commands isolates the section of HTML code we want
html_table2=soup2.find("pre")
#Extract the text from the code
text_table2=html_table2.get_text()
#split the text up into an array of lines
contents_array=text_table2.split('\n')
#Make a dataframe with the proper amount of space
df2=pd.DataFrame(columns=['start_date', 'start_time', 'end_date', 'end_time', 'start_frequency', 'end_frequency', 'flare_location', 'flare_region','flare_classification', 'cme_date', 'cme_time', 'cme_angle', 'cme_width', 'cme_speed','plots'],index = range(0,len(contents_array)-15))
#remove the first 12 lines, since we don't need them
contents_array=contents_array[12:]
#print(contents_array[0])
#get rid of the last 3 useless lines
del contents_array[-1]
del contents_array[-1]
del contents_array[-1]

##split up the array lines, and enter each entry into the proper column, since lines entries are split by spaces
##we also get rid of whitespace 
row_c=0
for row in contents_array:
    entry=row.split(' ')
    clean_entry=[]
    for thing in entry:
        if thing is None or thing =='':
            continue
        else:
            clean_entry.append(thing)
    col_c=0
    #inserting the cleaned entry into the dataframe
    for info in clean_entry:
        if col_c==15:
            continue
        #print(row_c,col_c,info)
        df2.iat[row_c, col_c] = info
        col_c+=1
    row_c+=1
    #break
```

```python
df2
```

```python

##Here we clean up the dataframe
df2=df2.replace('---','NaN')
df2=df2.replace('----','NaN')
df2=df2.replace('-----','NaN')
df2=df2.replace('------','NaN')
df2=df2.replace('--:--','NaN')
df2=df2.replace('--/--','NaN')
df2=df2.replace('24:00','23:59')
df2=df2.replace('????','NaN')

##We add a new column for when it has a halo
df2['Halo']="False"
##Populate the column
for index,row in df2.iterrows():
    if row["cme_angle"]=="Halo":
        row["Halo"]="True"

##clean up the old column        
df2=df2.replace('Halo','NA')
#print(df2)

##add a new column for defining if a CME is a lowerbound
df2['is_width_lower_bound']=""

##populate that column, and clean it of non-numeric symbols
for index, row in df2.iterrows():
    if '>' in row['cme_width']:
        row['is_width_lower_bound']="True"
        temp=row['cme_width'].strip('>')
        row['cme_width']=temp
    else:
        row['is_width_lower_bound']="False"
#print(df2)

## We want to combine our dates and times like we did last time
df2['start_datetime']="NaN"
df2['end_datetime']="NaN"
df2['cme_datetime']="NaN"

for index, row in df2.iterrows():
    temp=row['start_date'][0:5]
    #print(temp)
    if row['start_date']!="NaN" and row['start_time']!="NaN":
        #(row['start_date'],row['start_time'])
        row['start_datetime'] = (pd.to_datetime(row['start_date'] + ' ' + row['start_time']))
    if row['end_date']!="NaN" and row['end_time']!="NaN":
        #print(temp+row['end_date'],row['end_time'])
        row['end_datetime']= (pd.to_datetime(temp+row['end_date'] + ' ' + row['end_time']))
    if row['cme_date']!="NaN" and row['cme_time']!="NaN":
        #print(temp+row['cme_date'],row['cme_time'])
        row['cme_datetime']= (pd.to_datetime(temp+row['cme_date'] + ' ' + row['cme_time']))
        #row['cme_datetime'] = (pd.to_datetime(row['cme_date'] + ' ' + row['cme_time']))
        
##Now we want to drop the old date and time information        
df2=df2.drop('start_date',axis=1)
df2=df2.drop('start_time',axis=1)
df2=df2.drop('end_date',axis=1)
df2=df2.drop('end_time',axis=1)
df2=df2.drop('cme_date',axis=1)
df2=df2.drop('cme_time',axis=1)
```

```python
##Question 1##

#We do not get all the same data as is in the Top50 Flares website. 
#There must be extrasources of data used by the Top50 website to include flares not listed in the NASA data set

# Get rows with X class flare
topFlares = df2.loc[df2['flare_classification'].str.contains('X')]
topFlares.is_copy = False
# Get rid of X for sorting
topFlares['flare_classification'] = topFlares['flare_classification'].str.lstrip('X')
# Change type to float
topFlares['flare_classification'] = topFlares.flare_classification.astype(float)
# Sort by flare class
topFlares = topFlares.sort_values('flare_classification', ascending = False)
# Extract the top 50
topFlares = topFlares.head(50)
# Put back the X
topFlares['flare_classification'] = topFlares.flare_classification.astype(str)
topFlares['flare_classification'] = "X" + topFlares['flare_classification']
#topFlares
```

```python
##Question 3##

#    (a) a short description (2 sentences) of what the intent of your plot is (think in terms of our discussion on how we show variation, 
#    co-variation in terms of central trend, spread, skew etc.), 
#The intent is to visualize the difference between the top 50 and the rest. 
#With this visualizaiton we can see if more energentic CME produce halos more often.
#    (c)a short text description of your plot, and
#My plot shows the proportion of halo CMEs in top 50 vs halo CMEs in all the data.
#    (d) a sentence or two of interpretation of your plot (again think of variation, co-variation, etc.).
#The plot shows that the higher energy CME are more likely to have Halo CME versus the general CME. With over 80% of top 50 CME having halos and 
#only about half of general CME having halos

import seaborn as sns
import matplotlib.pyplot as plt

dfHalo = df2.loc[df2['Halo'].str.contains('T')]
dfHalo.is_copy = False
total_halos = dfHalo.shape[0]
total = df2.shape[0]
dfTopHalo = topFlares.loc[topFlares['Halo'].str.contains('T')]
dfTopHalo.is_copy = False
top_halos = dfTopHalo.shape[0]
top = topFlares.shape[0]
dfplot = pd.DataFrame({'Proportion has Halo': ['Top 50', 'All'], 'Proportion': [(top_halos/top), (total_halos/total)]})

sns.barplot(x = 'Proportion has Halo', y = 'Proportion', data=dfplot)
```
